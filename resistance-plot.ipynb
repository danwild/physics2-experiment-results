```python
import numpy as np
import matplotlib.pyplot as plt
from statistics import mean
```

```python
# ----------------------------------- RESISTOR 1 -------------
# resistor = '1'
# y axis (Current - AMPS)
# y_values = [0.05, 0.13, 0.17, 0.22] 
# x axis  (Electric potential - VOLTS)
# x_values = [0.0015, 0.0036, 0.0053, 0.0076] 

# ----------------------------------- RESISTOR 2 -------------
resistor = '2'
# # y axis (Current - AMPS)
y_values = [0.11, 0.22, 0.34, 0.41]
# # x axis  (Electric potential - VOLTS)
x_values = [0.0014, 0.0035, 0.0047, 0.0062] 
```

```python
# slope 1
slope1 = (0.22 - 0.05) / (0.0076 - 0.0015)
slope1
```

```python
# slope 2
(0.41 - 0.11) / (0.0062 - 0.0014)
```

```python
plt.figure(figsize=(20,10))

area = 200
plot_freqs = plt.scatter(x_values, y_values, s=area, marker="x", vmin=0)

plt.title('Resistor #' + resistor + ' - measured current vs electric potential', fontsize=30)
plt.xlabel('Electric potential (Volts)', fontsize=20)
plt.ylabel('Current (Amperes)', fontsize=20)

plt.ylim(bottom=0)
plt.xlim(left=0)

plt.tick_params(axis='x', labelsize=16)
plt.tick_params(axis='y', labelsize=16)
  
# line of best fit
x = np.insert(x_values, 0, 0)
# x = x_values

y = np.insert(y_values, 0, 0)
# y = y_values

slope, intercept = np.polyfit(x, y, 1)

# force intercept 0
intercept = 0
poly1d_fn = np.poly1d([slope, intercept]) 

plt.plot(x, np.array(poly1d_fn(x)), color='grey', alpha=0.3, label="Line of best fit")

plt.legend(fontsize=20, loc='lower right')
plt.show()
```
